```python
# Importando pacotes
import yfinance as yf
import pandas as pd
```

```python
def collect_price_tickets(start_date,end_date,path_tickets = '../data/bovespa.json',path_prices = '../data/prices.pickle'):
    # Lendo dados do indice bovespa
    df_tickets = pd.read_json(path_tickets)
    # Adicionando extensao .SA
    df_tickets['tickets'] = df_tickets['tickets'] + '.SA'
    # Transformando em lista
    list_tickets = df_tickets['tickets'].to_list()
    # Obtendo as cotacoes 
    df_prices = yf.download(list_tickets, start = start_date, end = end_date)
    # Salvando no disco apenas os precos
    df_prices['Adj Close'].to_csv(path_prices)  
```
